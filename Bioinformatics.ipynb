```python
# Example python code to generate condons or kmer
def generate_trimers():
    bases = ['A','U','G','C']
    result = []
    for base1 in bases:
        for base2 in bases:
            for base3 in bases:
                result.append( base1 + base2 + base3 )
    return result
rna=generate_trimers()
print(*rna)
```

def generate_kmers_rec(length):
    if length == 1:
        return ['A','U','G','C']
    else:
        result = []
        for seq in generate_kmers_rec( length - 1 ):
            for base in ['A','U','G','C']:

        return result
</code>

```python
# Using recursion, and create kmer of any length 
def generate_kmers_rec(length):
    if length == 1:
        return ['A','U','G','C']
    else:
        result = []
        for seq in generate_kmers_rec( length - 1 ):
            for base in ['A','U','G','C']:
                result.append( seq + base )
        return result

rna = generate_kmers_rec(3)
print( *rna )
```

# AI and Protein folding : Alphafold
AlphaFold is
a revolutionary artificial intelligence system developed by Google DeepMind that predicts a protein's 3D structure from its 1D amino acid sequence with near-experimental accuracy. By solving the 50-year-old "protein folding problem," AlphaFold has transformed structural biology, accelerating research in drug discovery, disease understanding, and biotechnology.
